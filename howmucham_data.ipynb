```
data = []
locations = []
amounts = []
for i in range(853):
  #Get the page
  URL = "https://nigeriapropertycentre.com/for-rent?page=" + str(i)
  requests.get(URL)

  #Fetch page content
  web_page = bs4.BeautifulSoup(requests.get(URL, {}).text, "lxml")

  #Get all the addresses and prices on the page
  address = web_page.find_all(name="address", attrs={"class": "voffset-bottom-10"})
  price = web_page.find_all(name ="span", attrs={"class": "price"})[1::1]

  #Get all locations
  for i in address:
    locations.append(i.strong.text)

  #Get their prices
  for i in price:
    amounts.append(i.text)

amounts = amounts[::2]

# #Put the gotten locations and their prices in a list
# for i in range(len(locations)):
#   data.append([locations[i], amounts[i]])

# #Tabulize my data
# data = pd.DataFrame(data)

# #View my data
# data.head()
```
